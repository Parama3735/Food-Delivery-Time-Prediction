### Cross-Validation for Model Stability

```python
from sklearn.model_selection import KFold
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

# Initialize KFold cross-validator
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# Lists to store metrics for each fold
mae_scores = []
mse_scores = []
rmse_scores = []
r2_scores = []

# Perform cross-validation
for train_index, val_index in kf.split(x):
    x_train_fold, x_val_fold = x.iloc[train_index], x.iloc[val_index]
    y_train_fold, y_val_fold = y.iloc[train_index], y.iloc[val_index]

    # Re-initialize StandardScaler for each fold to avoid data leakage
    sc_X_fold = StandardScaler()
    sc_y_fold = StandardScaler()

    x_train_scaled = sc_X_fold.fit_transform(x_train_fold)
    x_val_scaled = sc_X_fold.transform(x_val_fold)

    y_train_scaled = sc_y_fold.fit_transform(y_train_fold.values.reshape(-1, 1))
    # We transform y_val_fold but don't fit it to prevent data leakage
    y_val_scaled = sc_y_fold.transform(y_val_fold.values.reshape(-1, 1))

    # Train a new Linear Regression model for each fold
    lr_fold = LinearRegression()
    lr_fold.fit(x_train_scaled, y_train_scaled)
    y_pred_scaled = lr_fold.predict(x_val_scaled)

    # Inverse transform predictions to original scale for meaningful error calculation
    y_pred_original = sc_y_fold.inverse_transform(y_pred_scaled)

    # Calculate metrics for the current fold
    mae_scores.append(mean_absolute_error(y_val_fold, y_pred_original))
    mse_scores.append(mean_squared_error(y_val_fold, y_pred_original))
    rmse_scores.append(np.sqrt(mean_squared_error(y_val_fold, y_pred_original)))
    r2_scores.append(r2_score(y_val_fold, y_pred_original))

print("Cross-Validation Results:")
print(f"  Average MAE: {np.mean(mae_scores):.2f} (Std: {np.std(mae_scores):.2f})")
print(f"  Average MSE: {np.mean(mse_scores):.2f} (Std: {np.std(mse_scores):.2f})")
print(f"  Average RMSE: {np.mean(rmse_scores):.2f} (Std: {np.std(rmse_scores):.2f})")
print(f"  Average R-squared: {np.mean(r2_scores):.2f} (Std: {np.std(r2_scores):.2f})")
```

### Correcting Preprocessing and Training a New Model

To ensure accurate predictions, we need to correctly preprocess the data. The previous approach of reusing a single `LabelEncoder` across multiple columns (in cell `2klwPfjQM0O9`) caused issues with retaining the correct mappings for all categorical features.

We will now:
1.  Reload the raw data to start fresh with string categories.
2.  Re-apply the same NaN imputation and outlier handling steps.
3.  Create and fit *separate* `LabelEncoder` instances for each categorical column.
4.  Split and scale the data using `StandardScaler`.
5.  Train a *new* `LinearRegression` model (`lr_final`) with this correctly preprocessed data.

```python
# 1. Reload the raw data
raw_data_df = pd.read_csv("/content/Food_Delivery_Times.csv")

# 2. Re-apply NaN imputation (as originally done)
raw_data_df.loc[raw_data_df["Weather"].isnull(), "Weather"] = "Clear"
raw_data_df.loc[raw_data_df["Traffic_Level"].isnull(), "Traffic_Level"] = "Medium"
raw_data_df.loc[raw_data_df["Time_of_Day"].isnull(), "Time_of_Day"] = "Evening"
raw_data_df.loc[raw_data_df["Courier_Experience_yrs"].isnull(), "Courier_Experience_yrs"] = 3.0

# 3. Winsorize 'Delivery_Time_min' (as originally done)
from scipy.stats.mstats import winsorize
trim = 0.05
raw_data_df["Delivery_Time_min"] = winsorize(raw_data_df["Delivery_Time_min"], limits=trim)

# Drop 'Order_ID' as it's not a feature
raw_data_df_processed = raw_data_df.drop("Order_ID", axis=1)

# Separate features (x) and target (y)
x_processed = raw_data_df_processed.drop("Delivery_Time_min", axis=1)
y_processed = raw_data_df_processed["Delivery_Time_min"]

# Identify categorical columns for encoding
categorical_cols = ['Weather', 'Traffic_Level', 'Time_of_Day', 'Vehicle_Type']

# Store LabelEncoders in a dictionary
label_encoders = {}
for col in categorical_cols:
    le = LabelEncoder()
    # Fit on the unique values of the column in x_processed
    x_processed[col] = le.fit_transform(x_processed[col])
    label_encoders[col] = le

print("Data reloaded and preprocessed (NaNs, outliers, LabelEncoding).")
```

```python
# 4. Split data into training and testing sets
x_train_final, x_test_final, y_train_final, y_test_final = train_test_split(x_processed, y_processed, test_size=0.2, random_state=42)

# 5. Initialize and fit StandardScaler for features and target
sc_X_final = StandardScaler()
sc_y_final = StandardScaler()

# Scale features
x_train_scaled_final = sc_X_final.fit_transform(x_train_final)
x_test_scaled_final = sc_X_final.transform(x_test_final)

# Reshape and scale target variable
y_train_reshaped_final = y_train_final.values.reshape(-1, 1)
y_test_reshaped_final = y_test_final.values.reshape(-1, 1)
y_train_scaled_final = sc_y_final.fit_transform(y_train_reshaped_final)
y_test_scaled_final = sc_y_final.transform(y_test_reshaped_final)

print("Data split and scaled.")
```

```python
# 6. Train a new Linear Regression model with correctly scaled data
lr_final = LinearRegression()
lr_final.fit(x_train_scaled_final, y_train_scaled_final)

print("New Linear Regression model (lr_final) trained successfully with correctly preprocessed data.")
```
